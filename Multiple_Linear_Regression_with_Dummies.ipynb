# Multiple Linear Regression with Dummies

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns

# Configurar el estilo de las gr√°ficas con seaborn
sns.set()
```

## Load the data

```python
# Leer los datos desde el archivo CSV
raw_data = pd.read_csv('real_estate_price_size_year_view.csv')
```

```python
# Visualizar las primeras filas del DataFrame para entender su estructura
raw_data.head()
```

```python
# Obtener estad√≠sticas descriptivas del DataFrame
raw_data.describe(include='all')
```

## Create a dummy variable for 'view'

```python
# Copiar los datos para evitar modificar el conjunto de datos original
data = raw_data.copy()

# Mapear la variable 'view' a valores num√©ricos (0 para 'No sea view', 1 para 'Sea view')
data['view'] = data['view'].map({'Sea view': 1, 'No sea view': 0})
```

```python
# Visualizar las primeras filas del DataFrame modificado
data.head()
```

## Create the regression

### Declare the dependent and the independent variables

```python
# Definir la variable dependiente y las variables independientes
y = data['price']
x1 = data[['size', 'year', 'view']]
```

### Regression

```python
# A√±adir una columna de unos (constantes) a la matriz de caracter√≠sticas
x = sm.add_constant(x1)

# Ajustar un modelo de regresi√≥n lineal utilizando OLS (M√≠nimos Cuadrados Ordinarios)
results = sm.OLS(y, x).fit()

# Mostrar un resumen detallado de los resultados del modelo de regresi√≥n lineal
results.summary()
```
